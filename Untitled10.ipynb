```python
import csv
rows = []
with open('star_data_final.csv', 'r') as f:
    csv_r = csv.reader(f)
    for i in csv_r:
        rows.append(i)

headers = rows[0]
stars_data = rows[1:]
mass = rows[3]
headers[0] = 'Index'

print(headers)
print(mass)
```

```python
star_data = []
for star in stars_data:
    if star[3] != '?': 
        star[3] = float(star[3])*1.989e+30
    
    if star[4] != '?':
        star[4] = float(star[4])*6.957e+8
    star_data.append(star)

print(star_data)
```

```python
gravity_star = []
for star in star_data:
    if star[3] != '?' and star[4] != '?':
        gravity = (6.674e-11 * float(star[3]))/(float(star[4])*float(star[4]))
    star.append(gravity)
    gravity_star.append(star)

print(gravity_star)
```

```python
Name = []
Distance = []
Gravity = []
Mass = []
Radius = []


for data in gravity_star:
    Name.append(data[1])
    Distance.append(data[2])
    Gravity.append(data[5])
    Radius.append(data[4])
    Mass.append(data[3])

  
print(Name)
print(Distance)
print(Radius)
print(Gravity)
print(Mass)
```

```python
with open("new.csv","a+") as f:
    csvwriter = csv.writer(f)
    csvwriter.writerow(headers)
    csvwriter.writerows(stars_data)
```
